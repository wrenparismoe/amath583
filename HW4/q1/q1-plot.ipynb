```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("HW4/q1/performance.csv")

data["write_bandwidth"] = data["bytes"] / data["write_time"]
data["read_bancwidth"] = data["bytes"] / data["read_time"]
```

```python
# Set plot style to ggplot
plt.style.use("ggplot")

# Plot write and read bandwidths
fig = plt.figure(figsize=(10, 8))
# set title
plt.title("Write and Read Bandwidths")


plt.plot(data["n"], data["write_bandwidth"], label="write")
plt.plot(data["n"], data["read_bancwidth"], label="read")

plt.xlabel("n")
plt.ylabel("Bandidth (B/s)")
plt.legend()
```
