```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('results.csv')
data
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8, 5))
plt.plot(data['numThreads'], data['elapsed_time'], 'o-', label='Elapsed Time')
plt.xlabel('Number of Threads')
plt.ylabel('Elapsed Time (miliseconds)')
plt.title('Elapsed Time vs. Number of Threads')
plt.legend()
```

```python
dataError = pd.read_csv("resultsError.csv")
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(np.log(dataError['n']), dataError['error'], 'o-', label='Absolute Error')
plt.xlabel('n (log scale)')
plt.ylabel(r'Absolute Error $y_{true} - y_{approx}$')
plt.title('Absolute Error vs. n')
plt.legend()
```
