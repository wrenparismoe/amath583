```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("performance-best.csv", sep=',', header=0, names=["n", "Elapsed Time", "FLOPs"], dtype=np.float64)

window = 16
data["FLOPs Avg"] = data["FLOPs"].rolling(window).mean().shift(-int(window/2))
```

```python
plt.figure(figsize=(12,5))
plt.title(r"Level 1 BLAS Performance ($y = \alpha x + y$)")
sns.lineplot(
    x="n",
    y="FLOPs",
    data=data,
    label="FLOPs",
    linewidth=1,
)
sns.lineplot(
    x="n",
    y="FLOPs Avg",
    data=data,
    label=f"FLOPs Rolling Avg (window={window})",
    linewidth=2,
)
plt.xlabel("n", fontsize=12)
plt.ylabel("FLOPs", fontsize=12)
```
