```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("performance.csv", sep=',', header=0, names=['blocksize', 'Elapsed Time', 'FLOPs'], dtype=np.float64)
```

```python
plt.figure(figsize=(10,5))
plt.title("Level 1 BLAS Performance with Loop Unrolling")

sns.scatterplot(
	x="blocksize",
	y="FLOPs",
	data=data,
	label="FLOPs",
)

plt.xlabel("blocksize", fontsize=12)
plt.ylabel("FLOPs", fontsize=12)
```
