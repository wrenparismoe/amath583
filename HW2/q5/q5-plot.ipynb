```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
# %matplotlib inline
import seaborn as sns
```

```python
dataL1 = pd.read_csv("L1-performance.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

windowL1 = 14
dataL1["FLOPs Avg"] = dataL1["FLOPs"].rolling(windowL1).mean().shift(-int(windowL1/2))
```

```python
dataL2 = pd.read_csv("L2-performance.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

windowL2 = 10
dataL2["FLOPs Avg"] = dataL2["FLOPs"].rolling(windowL2).mean().shift(-int(windowL2/2))
```

```python
dataL3 = pd.read_csv("L3-performance.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)[:-1]

windowL3 = 10
dataL3["FLOPs Avg"] = dataL3["FLOPs"].rolling(windowL3).mean().shift(-int(windowL3/2))
```

```python
plt.style.use("ggplot")
# plt.style.use("Solarize_Light2")
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12,12))
fig.tight_layout(pad=4)
fig.subplots_adjust(top=0.93)
fig.suptitle("Template L1, L2, L3 BLAS Performance with `-O2` Optimization", fontsize=16)

# Level 1 BLAS
ax1.set_title(r"Level 1 BLAS Performance ($y=\alpha x + y$)", fontsize=12)
sns.lineplot(
    ax=ax1,
    x="n",
    y="FLOPs",
    data=dataL1,
    label="FLOPs",
    linewidth=1,
)
sns.lineplot(
    ax=ax1,
    x="n",
    y="FLOPs Avg",
    data=dataL1,
    label=f"FLOPs Rolling Avg (window={windowL1})",
    linewidth=2,
    # color="red",
)
ax1.set_xlabel("n", fontsize=12)
ax1.set_ylabel("FLOPs", fontsize=12)

# Level 2 BLAS
ax2.set_title(r"Level 2 BLAS Performance ($y=\alpha Ax + \beta y$)", fontsize=12)
sns.lineplot(
    ax=ax2,
    x="n",
    y="FLOPs",
    data=dataL2,
    label="FLOPs",
    linewidth=1,
)
sns.lineplot(
    ax=ax2,
    x="n",
    y="FLOPs Avg",
    data=dataL2,
    label=f"FLOPs Rolling Avg (window={windowL2})",
    linewidth=2,
    # color="red",
)
ax2.set_xlabel("n", fontsize=12)
ax2.set_ylabel("FLOPs", fontsize=12)

# Level 3 BLAS
ax3.set_title(r"Level 3 BLAS Performance ($y=\alpha AB + \beta C$)", fontsize=12)
sns.lineplot(
    ax=ax3,
    x="n",
    y="FLOPs",
    data=dataL2,
    label="FLOPs",
    linewidth=1.2,
)
sns.lineplot(
    ax=ax3,
    x="n",
    y="FLOPs Avg",
    data=dataL2,
    label=f"FLOPs Rolling Avg (window={windowL3})",
    linewidth=1.4,
    # color="red",
)
ax3.set_xlabel("n", fontsize=12)
ax3.set_ylabel("FLOPs", fontsize=12)
```

```python
plt.figure(figsize=(12,5))
plt.title(r"Level 1 Template BLAS Performance ($y=\alpha x + y$) with `-O2` Optimization")
sns.lineplot(
    x="n",
    y="FLOPs",
    data=dataL1,
    label="FLOPs",
    linewidth=1,
)
sns.lineplot(
    x="n",
    y="FLOPs Avg",
    data=dataL1,
    label=f"FLOPs Rolling Avg (window={windowL1})",
    linewidth=2,
)
plt.xlabel("n", fontsize=12)
plt.ylabel("FLOPs", fontsize=12)
```

```python
plt.figure(figsize=(12,5))
plt.title(r"Level 2 Template BLAS Performance ($y=\alpha Ax + \beta y$) with `-O2` Optimization")
sns.lineplot(
    x="n",
    y="FLOPs",
    data=dataL2,
    label="FLOPs",
    linewidth=1,
)
sns.lineplot(
    x="n",
    y="FLOPs Avg",
    data=dataL2,
    label=f"FLOPs Rolling Avg (window={windowL2})",
    linewidth=2,
)
plt.xlabel("n", fontsize=12)
plt.ylabel("FLOPs", fontsize=12)
```

```python
plt.figure(figsize=(12,5))
plt.title(r"Level 3 Template BLAS Performance ($y=\alpha AB + \beta C$) with `-O2` Optimization")
sns.lineplot(
    x="n",
    y="FLOPs",
    data=dataL2,
    label="FLOPs",
    linewidth=1,
)
sns.lineplot(
    x="n",
    y="FLOPs Avg",
    data=dataL2,
    label=f"FLOPs Rolling Avg (window={windowL3})",
    linewidth=2,
)
plt.xlabel("n", fontsize=12)
plt.ylabel("FLOPs", fontsize=12)
```
