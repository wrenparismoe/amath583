```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("performance.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window = 3
data["FLOPs Avg"] = data["FLOPs"].rolling(window).mean().shift(-int(window/2))
```

```python
plt.style.use("ggplot")
# plt.style.use("Solarize_Light2")
fig = plt.figure(figsize=(12,5))
fig.tight_layout(pad=4)
plt.title(r"Strassen MM Performance with Base Case $n=1$ and `-O3` Optimization", fontsize=14)
sns.lineplot(
    x="n",
    y="FLOPs",
    data=data,
    label="FLOPs",
    linewidth=1,
)
# sns.lineplot(
#     x="n",
#     y="FLOPs Avg",
#     data=data,
#     label=f"FLOPs Rolling Avg (window={window})",
#     linewidth=2,
# )
plt.xlabel("n", fontsize=12)
plt.ylabel("FLOPs", fontsize=12)
```
