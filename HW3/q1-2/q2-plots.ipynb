```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
# %matplotlib inline
# import seaborn as sns
```

```python
data_jki_base = pd.read_csv("jki-performance-base.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_jki = 10
data_jki_base["FLOPs Avg"] = data_jki_base["FLOPs"].rolling(window_jki).mean().shift(-int(window_jki/2))
```

```python
data_kij_base = pd.read_csv("kij-performance-base.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_kij = 10
data_kij_base["FLOPs Avg"] = data_kij_base["FLOPs"].rolling(window_kij).mean().shift(-int(window_kij/2))
```

```python
data_jki_O3 = pd.read_csv("jki-performance-O3.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_jki_O3 = 10
data_jki_O3["FLOPs Avg"] = data_jki_O3["FLOPs"].rolling(window_jki_O3).mean().shift(-int(window_jki_O3/2))
```

```python
data_kij_O3 = pd.read_csv("kij-performance-O3.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_kji_O3 = 10
data_kij_O3["FLOPs Avg"] = data_kij_O3["FLOPs"].rolling(window_kji_O3).mean().shift(-int(window_kji_O3/2))
```

```python
data_jki_fmath = pd.read_csv("jki-performance-fmath.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_jki_fmath = 10
data_jki_fmath["FLOPs Avg"] = data_jki_fmath["FLOPs"].rolling(window_jki_fmath).mean().shift(-int(window_jki_fmath/2))
```

```python
data_kij_fmath = pd.read_csv("kij-performance-fmath.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_kij_fmath = 10
data_kij_fmath["FLOPs Avg"] = data_kij_fmath["FLOPs"].rolling(window_kij_fmath).mean().shift(-int(window_kij_fmath/2))
```

```python
data_jki_O3_march = pd.read_csv("jki-performance-O3-march.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_jki_O3_march = 10
data_jki_O3_march["FLOPs Avg"] = data_jki_O3_march["FLOPs"].rolling(window_jki_O3_march).mean().shift(-int(window_jki_O3_march/2))
```

```python
data_kij_O3_march = pd.read_csv("kij-performance-O3-march.csv", sep=',', header=0, names=['n', 'Time', 'FLOPs'], dtype=np.float64)

window_kji_O3_march = 10
data_kij_O3_march["FLOPs Avg"] = data_kij_O3_march["FLOPs"].rolling(window_kji_O3_march).mean().shift(-int(window_kji_O3_march/2))
```

```python
plt.style.use("ggplot")
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16,9))
fig.tight_layout(pad=4)
fig.subplots_adjust(top=0.92)
plt.suptitle(r"Double Precision L3 BLAS Performance for Loop Permutations {jki} and {kij}  ($y=\alpha AB + \beta C$)", fontsize=16)

# No Optimization Plot
ax1.set_title("No Optimization", fontsize=12)
ax1.scatter(x=data_jki_base["n"], y=data_jki_base["FLOPs"], label="{jki} FLOPs", marker='.', facecolors='none', edgecolors='r')
ax1.plot(data_jki_base["n"], data_jki_base["FLOPs Avg"], label="{jki} FLOPs Rolling Avg", linewidth=1.5, color='orange')

ax1.scatter(x=data_kij_base["n"], y=data_kij_base["FLOPs"], label="{kij} FLOPs", marker='.', facecolors='none', edgecolors='b')
ax1.plot(data_kij_base["n"], data_kij_base["FLOPs Avg"], label="{kij} FLOPs Rolling Avg", linewidth=1.5,color='cyan')

ax1.set_xlabel("n", fontsize=12)
ax1.set_ylabel("FLOPs", fontsize=12)
ax1.legend(loc='lower right', fontsize=10)

# O3 Optimization Plot
ax2.set_title("`-O3` Optimization", fontsize=12)
ax2.scatter(x=data_jki_O3["n"], y=data_jki_O3["FLOPs"], label="{jki} FLOPs", marker='.', facecolors='none', edgecolors='r')
ax2.plot(data_jki_O3["n"], data_jki_O3["FLOPs Avg"], label="{jki} FLOPs Rolling Avg", linewidth=1.5, color='orange')

ax2.scatter(x=data_kij_O3["n"], y=data_kij_O3["FLOPs"], label="{kji} FLOPs", marker='.', facecolors='none', edgecolors='b')
ax2.plot(data_kij_O3["n"], data_kij_O3["FLOPs Avg"], label="{kji} FLOPs Rolling Avg", linewidth=1.5,color='cyan')

ax2.set_xlabel("n", fontsize=12)
ax2.set_ylabel("FLOPs", fontsize=12)
ax2.legend(loc='lower right', fontsize=10)

# ffast-math Optimization Plot
ax3.set_title("`-ffast-math` Optimization", fontsize=12)
ax3.scatter(x=data_jki_fmath["n"], y=data_jki_fmath["FLOPs"], label="{jki} FLOPs", marker='.', facecolors='none', edgecolors='r')
ax3.plot(data_jki_fmath["n"], data_jki_fmath["FLOPs Avg"], label="{jki} FLOPs Rolling Avg", linewidth=1.5, color='orange')

ax3.scatter(x=data_kij_fmath["n"], y=data_kij_fmath["FLOPs"], label="{kij} FLOPs", marker='.', facecolors='none', edgecolors='b')
ax3.plot(data_kij_fmath["n"], data_kij_fmath["FLOPs Avg"], label="{kij} FLOPs Rolling Avg", linewidth=1.5,color='cyan')

ax3.set_xlabel("n", fontsize=12)
ax3.set_ylabel("FLOPs", fontsize=12)
ax3.legend(loc='lower right', fontsize=10)

# O3 -march=native Optimization Plot
ax4.set_title("`-O3 -march=native` Optimization", fontsize=12)
ax4.scatter(x=data_jki_O3_march["n"], y=data_jki_O3_march["FLOPs"], label="{jki} FLOPs", marker='.', facecolors='none', edgecolors='r')
ax4.plot(data_jki_O3_march["n"], data_jki_O3_march["FLOPs Avg"], label="{jki} FLOPs Rolling Avg", linewidth=1.5, color='orange')

ax4.scatter(x=data_kij_O3_march["n"], y=data_kij_O3_march["FLOPs"], label="{kij} FLOPs", marker='.', facecolors='none', edgecolors='b')
ax4.plot(data_kij_O3_march["n"], data_kij_O3_march["FLOPs Avg"], label="{kij} FLOPs Rolling Avg", linewidth=1.5,color='cyan')

ax4.set_xlabel("n", fontsize=12)
ax4.set_ylabel("FLOPs", fontsize=12)
ax4.legend(loc='lower right', fontsize=10)
```
